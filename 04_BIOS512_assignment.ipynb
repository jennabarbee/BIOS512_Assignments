Use `read_csv` to load the data from the TidyTuesday GitHub repository. The project URL is can be found at [https://github.com/rfordatascience/tidytuesday/blob/master/data/2020/2020-10-27/](https://github.com/rfordatascience/tidytuesday/blob/master/data/2020/2020-10-27/).

```r
library('tidyverse')
```

```r
wind_turbines <- read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-10-27/wind-turbine.csv',)
```

Each row in the table represents a single turbine.

```r
wind_turbines |> head()
```

Make a bar chart 📊 **sorted by frequency** of turbine count per province territory. (Hint: `fct_infreq` will sort your factor levels by frequency for subsequent plotting)

```r
options(repr.plot.width = 6.25, repr.plot.height = 3.75, repr.plot.res = 200)

p = ggplot(wind_turbines, aes(y = province_territory |> fct_infreq() |> fct_rev())) +
    geom_bar(fill = 'darkseagreen') +
    labs(x="Number of Wind Turbines", 
              y="Province/Territory",
              title="Wind Turbines per Province/Territory",
              subtitle="Canada") +
    theme(plot.title=element_text(hjust=0.5),
          plot.subtitle=element_text(hjust=0.5))

p
```

Make a bar chart **sorted by frequency** of turbine count per manufacturer. **Group the low frequency manufacturers into an `Other` category**.

**Bonus:** Can you force the `Other` category to be the last bar? 

```r
options(repr.plot.width = 5, repr.plot.height = 3.75, repr.plot.res = 200)

p2 = ggplot(wind_turbines, aes(y = manufacturer |> fct_lump(n = 6) |> fct_infreq() |> 
                               fct_relevel('Other', after = Inf) |> fct_rev())) +
    geom_bar(fill = 'darkseagreen') +
    labs(x="Number of Wind Turbines", 
              y="Manufacturer",
              title="Wind Turbines per Manufacturer",
              subtitle="Canada") +
    theme(plot.title=element_text(hjust=0.5),
          plot.subtitle=element_text(hjust=0.5))

p2
```
